> # **LOGISTIC REGRESSION**



### **DEFINITION:**


*  Logistic regression is a powerful supervised Machine Learning algorithm used for binary classification problems.


*   It is based on Sigmoid function.


*   Logistic regression can used to solve problems efficiently when the target variable is categorical.


### **ADVANTAGES:**


*   Logistic regression is easier to implement, interpret, and very efficient to train.
*   Logistic regression is less inclined to over-fitting on simple models.

### **DISADVANTAGES:**


*   On high dimensional datasets, It may lead to the model being over-fit on the training set.
*   It is difficult to capture complex relationships using logistic regression.







```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000, random_state = 252)
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
model.predict_proba(X_test)
```

# ACCURACY AND EVALUATION OF THE MODEL

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

> # **DECISION TREE**



### **DEFINITION:**


*  Decision tree is a type of superivised Machine learning Algorithm, which can be used to solve problems with predefined Target Variable.


*   Decision Tree is fundamentally based on Entropy.


*   It can be used to solve for both categoricial and contiunous input and output variables. 


### **ADVANTAGES:**


*   Decision Trees are robust to outliers.
*   Decision trees can handle both classification and regression problems.

### **DISADVANTAGES:**


*   Overfitting is most practical difficulty in decision trees.
*   Sometimes the decision trees can create very complex trees which do not generalise the data well.







```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth= 2)
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model);
```
